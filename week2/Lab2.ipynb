```python
#cau 2
def phanloai_T1(x):
    if x < 5:
        return 'Kem'
    elif x < 7:
        return 'TB'
    elif x < 8:
        return 'Kha'
    else:
        return 'Gioi'

dftuyensinh['phanloai1'] = dftuyensinh['T1'].apply(phanloai_T1)
```

```python
#cau 3
bang_tanso = dftuyensinh['phanloai1'].value_counts()
bang_tanso
```

```python
# cau 4
plt.figure(figsize=(10,5))

for loai in dftuyensinh['phanloai1'].unique():
    data = dftuyensinh[dftuyensinh['phanloai1'] == loai]
    count_T1 = data['T1'].value_counts().sort_index()
    
    plt.plot(
        count_T1.index,
        count_T1.values,
        marker='o',
        label=loai
    )

plt.xlabel('T1')
plt.ylabel('Count')
plt.title('Multiple Line Chart - T1 theo phân loại phanloai1')
plt.legend(title='phanloai1')
plt.grid(True)
plt.show()
```

```python
#cau 5
plt.figure(figsize=(10,5))

for loai in dftuyensinh['phanloai1'].unique():
    data = dftuyensinh[dftuyensinh['phanloai1'] == loai]
    count_T1 = data['T1'].value_counts().sort_index()
    
    plt.vlines(
        x=count_T1.index,
        ymin=0,
        ymax=count_T1.values,
        label=loai
    )
    plt.plot(count_T1.index, count_T1.values, 'o')

plt.xlabel('T1')
plt.ylabel('Count')
plt.title('Drop-line Chart - T1 theo phân loại phanloai1')
plt.legend(title='phanloai1')
plt.grid(True)
plt.show()
```
