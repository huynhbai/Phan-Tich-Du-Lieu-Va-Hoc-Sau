```python
#câu 9
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("titanic_disaster.csv")
df['familySize'] = 1 + df['SibSp'] + df['Parch']
df['Alone'] = df['familySize'].apply(lambda x: 1 if x == 1 else 0)

df[['SibSp', 'Parch', 'familySize', 'Alone']].head()
```

```python
#câu 10
df['Cabin'] = df['Cabin'].fillna('Unknown')
df['typeCabin'] = df['Cabin'].apply(lambda x: x[0])

df[['Cabin', 'typeCabin']].head(10)
```

```python
#cau 12 
df['Survived_label'] = df['Survived'].astype(str).map({
    '0': 'Thiệt mạng',
    '1': 'Sống sót'
})

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sex', hue='Survived_label')
plt.title("Tương quan sống sót và thiệt mạng theo giới tính")
plt.xlabel("Giới tính")
plt.ylabel("Số lượng hành khách")
plt.show()

```

```python
#câu 13
df['Pclass_label'] = df['Pclass'].astype(str)

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Pclass_label', hue='Survived_label')
plt.title("Tỷ lệ sống sót theo hạng hành khách")
plt.xlabel("Hạng hành khách")
plt.ylabel("Số lượng hành khách")
plt.show()
```

```python
# CÂU 14

import pandas as pd

df['Agegroup'] = pd.cut(
    df['Age'],
    bins=[0,12,18,60,100],
    labels=['Kid','Teen','Adult','Older']
)

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Agegroup', hue='Survived_label')
plt.title("Sống sót theo nhóm tuổi")
plt.xlabel("Nhóm tuổi")
plt.ylabel("Số lượng hành khách")
plt.show()
```

```python
# CÂU 15
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='Alone', y='Survived')
plt.title("Xác suất sống sót theo việc đi một mình")
plt.xlabel("Đi một mình (1 = Có)")
plt.ylabel("Xác suất sống sót")
plt.ylim(0,1)
plt.show()
```

```python
# CÂU 16

plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Survived_label', y='Fare')
plt.title("Phân bố giá vé theo tình trạng sống sót")
plt.xlabel("Tình trạng")
plt.ylabel("Giá vé")
plt.show()
```

```python
# CÂU 17

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Embarked', hue='Survived_label')
plt.title("Sống sót theo cảng đến")
plt.xlabel("Cảng đến")
plt.ylabel("Số lượng hành khách")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Embarked', hue='Pclass_label')
plt.title("Phân bố hạng vé theo cảng đến")
plt.xlabel("Cảng đến")
plt.ylabel("Số lượng hành khách")
plt.show()
```
